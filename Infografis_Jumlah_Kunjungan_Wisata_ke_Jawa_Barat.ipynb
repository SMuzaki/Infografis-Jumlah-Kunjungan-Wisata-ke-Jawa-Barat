##**Distribusi Data Dengan Histogram**

```python
plt.figure(figsize=(10, 6))
data['jumlah_pengunjung'].hist(bins=30)
plt.title('Histogram of Jumlah Pengunjung')
plt.xlabel('jumlah_pengunjung')
plt.ylabel('Frequency')
plt.show()
```

##**Outliers dengan boxplot**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['jumlah_pengunjung'])
plt.title('Boxplot of Jumlah Pengunjung')
plt.show()
```
